```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
from numpy.core.fromnumeric import reshape
df=pd.read_csv('/content/drive/MyDrive/캡스톤/데이터만지기/refined_data')

scaler=MinMaxScaler()
np_df=np.asarray(df['log-view'])
np_df=np_df.reshape(-1,1)

scaler.fit(np_df)
df['normalized_view']=scaler.transform(np_df)

df.to_csv('/content/drive/MyDrive/캡스톤/데이터만지기/refined_data',header=True,index=True)
```

```python
import os
import numpy as np
file_list = os.listdir('/content/drive/MyDrive/캡스톤/data') # 무신사 데이터 이름 모두 file_list 로 불러들이기 
num_list=list()
path_list=list() # 저장할 사진 경로


for idx in range(1,16852): # 파일 이름은 str 이기 때문에 정수 idx 를 글자로 치환  
  idx_str=str(idx) 
  num_list.append(idx_str)

print(len(num_list))
```

```python
for idx in num_list: # 16000 개 리스트 중에
  if idx in file_list: # 만약 idx가 존재하면 
    path_list.append('/content/drive/MyDrive/캡스톤/data' +"/" +idx) #사진 경로를 부르고 
  else :
    path_list.append(np.nan) # 아니면 NaN 로 저장하라
path_list

print(len(path_list))
```

```python

df.loc[:,"img_path"]=path_list  #이걸 저장하라 

df
df.to_csv("/content/drive/MyDrive/캡스톤"+'/label_and_path',header=True,index=True)
```
